```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(
    {
        "name": ["john", "mary", "peter", "jeff", "bill", "lisa", "jose"],
        "age": [23, 78, 22, 19, 45, 33, 20],
        "gender": ["M", "F", "M", "M", "M", "F", "M"],
        "state": [
            "california",
            "dc",
            "california",
            "dc",
            "california",
            "texas",
            "texas",
        ],
        "num_children": [2, 0, 0, 3, 2, 1, 4],
        "num_pets": [5, 1, 0, 5, 2, 2, 3],
    }
)
df
```

```python
df.plot(kind="scatter", x="num_children", y="age", color="green")
plt.show()
```

```python
df.plot(kind="bar", x="name", y="num_pets", color="green")
plt.show()
```

```python
ax = plt.gca()
df.plot(kind="line", x="name", y="num_children", ax=ax)
df.plot(kind="line", x="name", y="num_pets", ax=ax)
plt.savefig("mygraph.png")
```

```python
df.plot(kind="pie", x="name", y="num_pets")
plt.show()
```

```python
labels = "john", "mary", "peter", "jeff", "bill", "lisa", "jose"
df.plot(kind="pie", x="name", y="num_pets", labels=labels)
plt.show()
```

```python
labels = "john", "mary", "peter", "jeff", "bill", "lisa", "jose"
df.plot(kind="pie", x="name", y="num_pets", labels=labels, legend=False)
plt.show()
```

```python
labels = "john", "mary", "peter", "jeff", "bill", "lisa", "jose"
df.plot(
    kind="pie",
    x="name",
    y="num_pets",
    labels=labels,
    legend=False,
    table=df,
    figsize=(8, 8),
    explode=(0, 0.2, 0, 0, 0, 0, 0),
)
plt.show()
```

```python
ax = df[["age"]].plot(kind="hist", bins=[0, 20, 40, 60, 80, 100], rwidth=0.9)
ax.set(title="Superman", xlabel="Range of ages", ylabel="Total amount")
plt.show()
```
