```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("project.csv")
belgium = df[df.Country == "Belgium"]
austria = df[df.Country == "Austria"]
```

```python
male_count = austria[austria.Gender == "Male"].Gender.count()
female_count = austria[austria.Gender == "Female"].Gender.count()
df = pd.DataFrame(
    {
        "Gender": ["Male"],
        "Male": [male_count],
    }
)
df.plot(kind="hist")
plt.show()

df = pd.DataFrame(
    {
        "Gender": ["Female"],
        "Female": [female_count],
    }
)
df.plot(kind="hist")
plt.show()
```
